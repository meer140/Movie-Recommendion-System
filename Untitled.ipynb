```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
movies.head(1)
```

```python
credits.head(1)
```

```python
credits.drop('title',axis=1,inplace=True)
```

```python
movies = pd.concat([movies,credits],axis=1)
```

```python
movies.head(1)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(1)
```

```python
movies.dropna(axis=0,inplace=True)
```

```python
movies.columns
```

```python
movies.shape
```

```python
movies.genres.apply(ast.literal_eval)[0]
```

```python
def convert(obj): 
    list=[]
    for i in obj.apply(ast.literal_eval):
        inner_list = []
        for j in i:
            inner_list.append(j["name"])
        list.append(inner_list)
    return list
        
    
```

```python
movies['genres'] = convert(movies['genres'])
```

```python
movies.head(1)
```

```python
movies.keywords = convert(movies['keywords'])
```

```python
movies.head()
```

```python
def convert_cast(obj): 
    list=[]
    for i in obj.apply(ast.literal_eval):
        inner_list = []
        count = 0
        for j in i:
            if(count < 3):
                inner_list.append(j["name"])
            else:
                break
            count=count+1;
            
        list.append(inner_list)
    return list
```

```python
movies.cast = convert_cast(movies.cast)
```

```python
movies.head(1)
```

```python
def convert_crew(obj): 
    list=[]
    for i in obj.apply(ast.literal_eval):
        inner_list = []
        for j in i:
            if(j["job"]=="Director"):
                inner_list.append(j["name"])
        list.append(inner_list)
    return list
```

```python
movies.crew = convert_crew(movies.crew)
```

```python
movies.head(1)
```

```python
movies.overview = movies.overview.apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tag']]
new_df
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    list = []
    for i in text.split():
        list.append(ps.stem(i))
        
    return " ".join(list)
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
new_df.tag[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
vectors
```

```python
cv.get_feature_names_out()[30:100]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_idx = new_df[new_df['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x : x[1])

    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
